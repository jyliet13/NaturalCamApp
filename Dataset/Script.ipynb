```python
import os
import shutil
from collections import defaultdict

# --- Rutas de tus datasets actuales ---
# ASEGÚRATE de que estas rutas sean CORRECTAS en tu sistema.
# Puedes copiar la ruta desde PyCharm (clic derecho en la carpeta -> Copy Path/Reference -> Path From Content Root)
OLD_DATASET_BASE = os.path.join('Fossil/Geo Fossils-I Dataset') # Las 6 clases originales
NEW_DATASET_BASE = os.path.join('Test/dataset_organizado')    # Las nuevas y algunas originales

```

```python
# --- Ruta del nuevo dataset unificado ---
# Aquí se copiarán todas las imágenes, organizadas por clase.
OUTPUT_DATASET_BASE = 'Dataset_Fosiles_Completo'
OUTPUT_ALL_IMAGES_DIR = os.path.join(OUTPUT_DATASET_BASE, 'all_images') # Todas las imágenes aquí, sin split inicial

def unify_datasets(old_base, new_base, output_base):
    """
    Unifica imágenes de dos datasets en una nueva estructura de carpetas por clase.
    Maneja duplicados añadiendo un sufijo.
    """
    print(f"Iniciando unificación de datasets...")
    print(f"Origen 1: {old_base}")
    print(f"Origen 2: {new_base}")
    print(f"Destino: {output_base}")

    # Asegura que el directorio de salida exista y que all_images también exista
    os.makedirs(output_base, exist_ok=True)
    os.makedirs(OUTPUT_ALL_IMAGES_DIR, exist_ok=True)

    copied_files_count = 0
    skipped_duplicates_count = 0
    error_count = 0

    # Diccionario para contar archivos copiados por clase y manejar duplicados
    file_counts_per_class = defaultdict(int)

    # Procesar el primer dataset (Geo Fossils-I Dataset)
    print(f"\nProcesando el dataset original: {old_base}")
    for class_name in os.listdir(old_base):
        class_path = os.path.join(old_base, class_name)
        if os.path.isdir(class_path):
            output_class_path = os.path.join(OUTPUT_ALL_IMAGES_DIR, class_name)
            os.makedirs(output_class_path, exist_ok=True) # Crea la carpeta para la clase en el destino

            print(f"  Copiando imágenes de la clase '{class_name}' (Origen 1)...")
            for filename in os.listdir(class_path):
                source_file_path = os.path.join(class_path, filename)
                if os.path.isfile(source_file_path):
                    try:
                        # Manejar posibles duplicados con sufijo
                        destination_file_path = os.path.join(output_class_path, filename)
                        original_filename_no_ext, ext = os.path.splitext(filename)

                        # Si el archivo ya existe, añadir un sufijo
                        if os.path.exists(destination_file_path):
                            file_counts_per_class[class_name] += 1
                            new_filename = f"{original_filename_no_ext}_{file_counts_per_class[class_name]}{ext}"
                            destination_file_path = os.path.join(output_class_path, new_filename)
                            skipped_duplicates_count += 1
                            # print(f"    Duplicado detectado: {filename}. Copiando como {new_filename}")

                        shutil.copy2(source_file_path, destination_file_path)
                        copied_files_count += 1
                    except Exception as e:
                        print(f"    Error copiando {source_file_path}: {e}")
                        error_count += 1

    # Resetear el contador para el segundo dataset, ya que los nombres podrían ser diferentes
    file_counts_per_class = defaultdict(int)

    # Procesar el segundo dataset (dataset_organizado)
    print(f"\nProcesando el nuevo dataset: {new_base}")
    for class_name in os.listdir(new_base):
        class_path = os.path.join(new_base, class_name)
        if os.path.isdir(class_path):
            output_class_path = os.path.join(OUTPUT_ALL_IMAGES_DIR, class_name)
            os.makedirs(output_class_path, exist_ok=True) # Crea la carpeta para la clase en el destino

            print(f"  Copiando imágenes de la clase '{class_name}' (Origen 2)...")
            for filename in os.listdir(class_path):
                source_file_path = os.path.join(class_path, filename)
                if os.path.isfile(source_file_path):
                    try:
                        # Manejar posibles duplicados con sufijo
                        destination_file_path = os.path.join(output_class_path, filename)
                        original_filename_no_ext, ext = os.path.splitext(filename)

                        # Si el archivo ya existe, añadir un sufijo
                        if os.path.exists(destination_file_path):
                            file_counts_per_class[class_name] += 1
                            new_filename = f"{original_filename_no_ext}_{file_counts_per_class[class_name]}{ext}"
                            destination_file_path = os.path.join(output_class_path, new_filename)
                            skipped_duplicates_count += 1
                            # print(f"    Duplicado detectado: {filename}. Copiando como {new_filename}")

                        shutil.copy2(source_file_path, destination_file_path)
                        copied_files_count += 1
                    except Exception as e:
                        print(f"    Error copiando {source_file_path}: {e}")
                        error_count += 1

    print(f"\n--- Resumen de Unificación ---")
    print(f"Total de archivos copiados: {copied_files_count}")
    if skipped_duplicates_count > 0:
        print(f"Archivos con nombres duplicados (se añadió sufijo): {skipped_duplicates_count}")
    if error_count > 0:
        print(f"Errores encontrados durante la copia: {error_count}")
    print(f"Dataset unificado creado en: {OUTPUT_ALL_IMAGES_DIR}")
    print(f"Puedes inspeccionar esta carpeta para ver todas las imágenes por clase.")
    print(f"Luego, podrás usar un script para dividir esto en Train/Validation/Test.")


```

```python
if __name__ == "__main__":
    # Ejecutar la función para unificar
    unify_datasets(OLD_DATASET_BASE, NEW_DATASET_BASE, OUTPUT_DATASET_BASE)
```

```python
import os

# Ruta al directorio donde están todas tus imágenes unificadas (con posibles duplicados _1, _2, etc.)
DATASET_PATH = 'Dataset_Fosiles_Completo/all_images'

```

```python
def delete_suffixed_duplicates(dataset_root_path):
    """
    Elimina archivos que parecen ser duplicados generados por el script de unificación,
    identificándolos por la presencia de un guion bajo seguido de un número y extensión
    (ej. 'imagen_1.jpg', 'foto_2.png').
    """
    print(f"Buscando y eliminando duplicados con sufijo '_' en: {dataset_root_path}")
    removed_count = 0

    for class_name in os.listdir(dataset_root_path):
        class_path = os.path.join(dataset_root_path, class_name)
        if os.path.isdir(class_path):
            print(f"Procesando clase: {class_name}")
            for filename in os.listdir(class_path):
                # Patrón para identificar los duplicados generados:
                # Busca un guion bajo seguido de uno o más dígitos, antes de la extensión.
                # Ejemplo: "imagen_1.jpg", "fossil_02.png"
                name_parts = os.path.splitext(filename)
                base_name = name_parts[0]
                extension = name_parts[1]

                # Comprueba si el nombre base termina con _ + número
                if '_' in base_name:
                    last_part = base_name.split('_')[-1]
                    if last_part.isdigit():
                        # Intenta eliminar el archivo
                        filepath_to_delete = os.path.join(class_path, filename)
                        try:
                            os.remove(filepath_to_delete)
                            removed_count += 1
                            # print(f"  Eliminado: {filepath_to_delete}") # Descomentar para depuración detallada
                        except Exception as e:
                            print(f"  Error al eliminar {filepath_to_delete}: {e}")

    print(f"\n--- Resumen de Eliminación de Duplicados por Nombre ---")
    print(f"Total de imágenes duplicadas (con sufijo '_') eliminadas: {removed_count}")
    print(f"Proceso completado en: {dataset_root_path}")


```

```python
if __name__ == "__main__":
    # Asegúrate de que DATASET_PATH apunte a la carpeta 'all_images'
    # Obtenemos la ruta absoluta para mayor robustez
    SCRIPT_DIR = os.path.dirname(os.path.abspath(__file__))
    FULL_DATASET_PATH = os.path.join(SCRIPT_DIR, DATASET_PATH)

    delete_suffixed_duplicates(FULL_DATASET_PATH)
```
